Exponentially Weighted Average

```python
def EWMA(df, column, beta = 0.9):

    lenght_ewma = df.shape[0]

    w_t = np.zeros(lenght_ewma)

    theta_t = df[column]

    for i in range(1, lenght_ewma):
        w_t[i] = beta*w_t[i-1] + (1-beta)*theta_t[i]

    return w_t.tolist()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Step 1: Filter data for 2019
data = data_all_years[data_all_years.year == 2019].copy()
data.reset_index(drop=True, inplace=True)

# Step 2: Drop invalid rows with unrealistic year, month, or day values
data = data[(data['year'] <= 2100) & (data['month'] <= 12) & (data['day'] <= 31)]

# Step 3: Create date column
data['date'] = pd.to_datetime(data[['year', 'month', 'day']])

# Step 4: Convert temp_f to temp_c (if not already done)
if 'temp_c' not in data.columns:
    data['temp_c'] = (data['temp_f'] - 32) * 5.0 / 9.0

# Step 5: Create year_day column if not exists
if 'year_day' not in data.columns:
    data.insert(0, 'year_day', range(1, data.shape[0] + 1))

# Step 6: Plotting
fig1, ax = plt.subplots(figsize=(15, 8))
x = data['date']

# Format the x-axis to show month names
myFmt = mdates.DateFormatter('%B')
ax.xaxis.set_major_formatter(myFmt)

# Set limits and plot
ax.set_xlim((x.min(), x.max()))
ax.plot(x, data['temp_c'], 'o', color='b', alpha=0.5, label='Temperature ($^\circ$C)')

# Labels and formatting
ax.set_title('Temperature Paris 2019')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature $^\circ$C')
ax.legend()
fig1.autofmt_xdate()  # Rotate date labels for better readability

plt.show()

# Ensure save directory exists
os.makedirs('../plots', exist_ok=True)

# Save figure
fig1.savefig('../plots/plot_1.png', dpi=300)

```

```python

data['EWA_0.9'] = EWMA(data, column='temp_c', beta = 0.9)
data['EWA_0.97'] = EWMA(data, column='temp_c', beta = 0.97)
data['EWA_0.5'] = EWMA(data, column='temp_c', beta = 0.5)
```

```python

days_vector = np.array(range(0, 91))
exp_days = np.power(0.9, days_vector)
```

```python
fig5, ax = plt.subplots(figsize=(13, 8))


ax.invert_xaxis()

plt.plot(days_vector, exp_days, 'o', alpha = 0.7, color = 'r', label = r"$\beta= 0.9$")

plt.title(r'Exponential weights: $( \beta^{90}, \beta^{89}, \beta^{88}, ...,\beta^{1}, 1)$')
plt.legend()
plt.xlabel("t")
plt.ylabel(r"$\beta^t$")

plt.show()

fig5.savefig('../plots/plot_5_1', dpi = 300)
```
